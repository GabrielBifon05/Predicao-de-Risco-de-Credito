Calculadora para simulação do risco de atraso ou sem atraso:

 - Utilizando o XGBoost, como melhor das 3 performances

```python
def prever_risco(xgb, input_dict, threshold=0.3):
    import pandas as pd

    # transformar em DataFrame
    df_input = pd.DataFrame([input_dict])

    # prever probabilidade
    prob = xgb.predict_proba(df_input)[0][1]

    # aplicar threshold
    classe = int(prob >= threshold)

    return prob, classe
```

```python
cliente = {
    'Unnamed: 0': 0,
    'RevolvingUtilizationOfUnsecuredLines': 0.8,
    'age': 35,
    'NumberOfTime30-59DaysPastDueNotWorse': 2,
    'DebtRatio': 1.5,
    'MonthlyIncome': 4000,
    'NumberOfOpenCreditLinesAndLoans': 5,
    'NumberOfTimes90DaysLate': 1,
    'NumberRealEstateLoansOrLines': 1,
    'NumberOfTime60-89DaysPastDueNotWorse': 0,
    'NumberOfDependents': 2
}

prob, classe = prever_risco(xgb, cliente)

print(f"Probabilidade de risco: {prob:.2f}")
print("Risco" if classe == 1 else "Sem risco")
```
